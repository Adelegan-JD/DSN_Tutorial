## 🏁 Mini-Project 1: Student Performance Analyzer
**Goal:** Combine everything to process a mini dataset.
1. Create a NumPy array of 5 scores.
2. Use a function to normalize those scores (scale to 0-1).
3. Put them in a Pandas DataFrame and calculate the mean score.

```python
# 1. NumPy Array
raw_scores = np.array([55, 92, 78, 40, 88])

```

```python
# 2. Normalization Function
def normalize(arr):
    return (arr - arr.min()) / (arr.max() - arr.min())

norm_scores = normalize(raw_scores)

```

```python
# 3. Pandas DataFrame
results = pd.DataFrame({
    'Raw': raw_scores,
    'Normalized': norm_scores
})

print("Student Performance:")
print(results)
print(f"\nAverage Normalized Score: {results['Normalized'].mean():.2f}")
```

## 🏁 Mini-Project 2: Sales Performance Analyzer
**Goal:** Carry out data analysis using Numpy and Pandas.
1. Create a NumPy array to simulate a company's sales dataset.
2. Put them in a Pandas DataFrame and calculate the mean score
3. Carry out various Pandas operations on it.


```python
# Create the dataset
import pandas as pd

data = {
    'Region': ['North', 'South', 'North', 'East', 'South'],
    'Sales': [1500, 2000, 1200, np.nan, 2500],
    'Expenses': [800, 1200, 900, 500, 1300]
}
df = pd.DataFrame(data)

```

```python
# 1. Inspection
print(df.info())
```

```python
# 2. Filling Missing Data
df['Sales'] = df['Sales'].fillna(df['Sales'].mean())
```

```python
# 3. New Column Creation
df['Profit'] = df['Sales'] - df['Expenses']
```

```python
# 4. Filtering Rows
print(df[df['Profit'] > 500])
```

```python
# 5. Label Indexing (.loc)
print(df.loc[0, 'Region'])
```

```python
# 6. Positional Indexing (.iloc)
print(df.iloc[0:2, :])
```

```python
# 7. Grouping (Sum by Region)
print(df.groupby('Region')['Sales'].sum())
```

```python

# 8. Sorting
print(df.sort_values(by='Sales', ascending=False))
```

```python
# 9. Counting Categorical Values
print(df['Region'].value_counts())
```

```python
# 10. Pivot Tables
print(df.pivot_table(values='Sales', index='Region', aggfunc='max'))
```

```python
# 11. Lambda Apply (Text Clean)
df['Region'] = df['Region'].apply(lambda x: x.upper())
```
