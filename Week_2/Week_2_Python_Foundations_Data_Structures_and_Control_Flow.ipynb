# Topic 1: Storing Data (Lists & Tuples)

### Everyday analogies
- **List = grocery list 🛒** (you can edit it)
- **Tuple = printed receipt 🧾** (fixed; you don’t edit it)

### Big Idea: Mutability
- **List = mutable** (changeable)
- **Tuple = immutable** (not changeable)

## 1.1 Lists — your grocery list 🛒

````python
# A list stores items in order (mutable)
Groceries = ["rice", "eggs", "tomatoes", "milk"]
print("Groceries:", Groceries)
print("Number of items:", len(Groceries))
````

### Indexing and slicing
- Indexing starts at **0**
- Negative indexing: **-1** means “last item”
- Slicing: `my_list[start:end]` (end is excluded)

````python
groceries = ["rice", "eggs", "tomatoes", "milk", "bread", "beans", "pepper"]

# Indexing
print("First item:", groceries[0])
print("Last item:", groceries[-1])

# Slicing
print("First 3 items:", groceries[0:3])
print("From index 2 to end:", groceries[2:])
print("from the first item to the third item", groceries[:3])
print("All items:", groceries[:])
# Start, stop, step

print(groceries[0:7:2])
````

### Mutability demo: editing a list

````python
groceries = ["rice", "eggs", "tomatoes", "milk"]
print("Before:", groceries)

groceries[2] = "pepper"      # change
groceries.append("bread")    # add to end
groceries.remove("eggs")     # remove by value

print("After:", groceries)
````

## 1.2 Tuples — your receipt 🧾

A **tuple** is like a list, but you **can’t change it** after creation.

Use tuples when:
- You want a fixed record (e.g., coordinates, dates, a profile)

````python
# A tuple is immutable
profile = ("Deborah", 20, "Nigeria")  # (name, age, country)
print(profile)

print("Name:", profile[0])
print("Age:", profile[1])
print("Country:", profile[2])
````

## 2.1 Dictionaries — your contact list 📱

A **dictionary** stores information as **key → value** pairs.

Example:
- key: `"Mom"`
- value: `"0803-123-4567"`

Why dictionaries are useful in everyday Python:
- Look up info by key (fast and readable)
- Build simple “systems” (contacts, inventory, settings, scores)

````python
contacts = {
    "Mom": "0803-123-4567",
    "Dad": "0802-111-2222",
    "Bestie": "0705-999-0000"
}

print("All contacts:", contacts)
print("Mom's number:", contacts["Mom"])
````

### Add, update, and delete dictionary entries

````python
contacts = {"Mom": "0803-123-4567", "Dad": "0802-111-2222"}
print("Before:", contacts)

contacts["Boss"] = "0901-555-1212"   # add
contacts["Mom"] = "0803-000-0000"    # update
del contacts["Dad"]                  # delete

print("After:", contacts)
````

### Dictionary methods: keys, values, items

````python
contacts = {"Mom": "0803-123-4567", "Boss": "0901-555-1212", "Dad": "0802-111-2222"}

print("Keys:", list(contacts.keys()))
print("Values:", list(contacts.values()))
print("Items:", list(contacts.items()))
````

## 2.2 Sets — your “no duplicates” basket 🧺

A **set** stores **unique** values (duplicates are automatically removed).

Why sets are useful in everyday Python:
- Remove duplicates (emails, usernames, repeated words)
- Check membership quickly (`in`)

````python
emails = ["a@x.com", "b@x.com", "a@x.com", "c@x.com", "b@x.com"]
unique_emails = set(emails)

print("Original list:", emails)
print("Unique set:", unique_emails)
print("Unique count:", len(unique_emails))
````

### Set operations (optional power tools)

````python
group_a = {"Ada", "Bola", "Chidi"}
group_b = {"Bola", "Dami", "Efe"}

print("Union (A or B):", group_a | group_b)
print("Intersection (A and B):", group_a & group_b)
print("Difference (A - B):", group_a - group_b)
print("Difference (B - A):", group_b - group_a)
````

# Hour 3 — Automation (Loops & Logic)

Automation is where Python starts to feel like superpowers:
- **for loops**: repeat for each item
- **while loops**: repeat while a condition is true
- **if/else**: make decisions

## 3.1 For loops — repeat for each item

````python
groceries = ["rice", "pepper", "bread"]
for item in groceries:
    print("Buy:", item)
````

````python
fruits = ['banana', 'apple', 'orange', 'guava', 'cherry']
for fruit in fruits:
    print(f'I love {fruit}') 
````

### Looping through dictionaries with .items()

````python
contacts = {"Mom": "0803-123-4567", "Boss": "0901-555-1212"}

for name, phone in contacts.items():
    print(f"{name} → {phone}")
````

### If/else inside loops (decision-making)

````python
numbers = [2, 7, 10, 13, 18]
for n in numbers:
    if n % 2 == 0:
        print(n, "is even ✅")
    else:
        print(n, "is odd 🔥")
````

# Final Lab (30–40 mins) — Inventory Management System 🧪

Build a simple console program that can:
1. View inventory
2. Add a new item
3. Update quantity
4. Sell an item
5. Exit

**Rules:**
- Use a **dictionary**: item → quantity
- Use a **while loop** for the menu
- Use **if/elif/else** for choices

````python
# Inventory Management System (Starter Template)
inventory = {"rice": 10, "beans": 5, "milk": 8}

while True:
    print("\n=== INVENTORY MENU ===")
    print("1) View inventory")
    print("2) Add new item")
    print("3) Update quantity")
    print("4) Sell item")
    print("5) Exit")

    choice = input("Choose an option (1-5): ").strip()

    if choice == "1":
        print("\n--- Current Inventory ---")
        for item, qty in inventory.items():
            print(f"{item}: {qty}")

    elif choice == "2":
        item = input("Enter item name: ").strip().lower()
        if item in inventory:
            print("Item already exists. Use 'Update quantity' instead.")
        else:
            qty_str = input("Enter quantity (number): ").strip()
            if qty_str.isdigit():
                inventory[item] = int(qty_str)
                print("Added ✅")
            else:
                print("Quantity must be a number ❌")

    elif choice == "3":
        item = input("Enter item name to update: ").strip().lower()
        if item not in inventory:
            print("Item not found ❌")
        else:
            qty_str = input("Enter new quantity: ").strip()
            if qty_str.isdigit():
                inventory[item] = int(qty_str)
                print("Updated ✅")
            else:
                print("Quantity must be a number ❌")

    elif choice == "4":
        item = input("Enter item name to sell: ").strip().lower()
        if item not in inventory:
            print("Item not found ❌")
        else:
            qty_str = input("Enter quantity to sell: ").strip()
            if qty_str.isdigit():
                qty_to_sell = int(qty_str)
                if qty_to_sell <= inventory[item]:
                    inventory[item] -= qty_to_sell
                    print("Sold ✅")
                else:
                    print("Not enough stock ❌")
            else:
                print("Quantity must be a number ❌")

    elif choice == "5":
        print("Goodbye 👋")
        break

    else:
        print("Invalid option. Choose 1–5.")
````
